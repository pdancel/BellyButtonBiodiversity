```python
#OTU IDs and Sample Values

#Read in Sample Data
sample_df = pd.read_csv('belly_button_biodiversity_samples.csv')
sample_df.sort_values(by='otu_id', ascending=False)
sample_df.head()
```

```python
sample_df_sort = sample_df.sort_values('BB_940', ascending=False)
sample_df1 = sample_df_sort[['otu_id', 'BB_940']]
sample_df1.head()
```

```python
#Filter Values greater than 1
sample_df_nan = sample_df1[sample_df1 > 0]
# Drop Null Values
sample_df_nan1 = sample_df_nan.dropna()
```

```python
sample_df_nan1
```

```python
#Create lists for OTU IDs and Sample Values
otu_id_list = sample_df_nan1['otu_id'].tolist()
sample_value_list = sample_df_nan1['BB_940'].tolist()
```

```python
samples_dict = [{'otu_ids': otu_id_list, 'sample_values': sample_value_list}]
print(samples_dict)
```
